Decision Tree Classification

Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
dataset = pd.read_csv('Iris.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
y
```

Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier()
model1.fit(X_train,y_train)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

Training the Decision Tree Classification model on the Training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

Predicting a new result

```python
print(classifier.predict(sc.transform([[5.9,3,5.1,1.8]])))
```

Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

Visualising the Training set results

```python
# Plotting
from sklearn.tree import plot_tree
from sklearn import tree
f_n = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
c_n = ["Setosa", "Versicolor", "Virginica"]
fig,ax=plt.subplots(nrows=1,ncols=1,dpi=300)
tree.plot_tree(model1,feature_names=f_n,class_names=c_n,filled=True)
fig.savefig('IrisTree.png')
```
